```python
def convert_date(input_DOB):
    year, month, date = input_DOB.split("-")
    return ("{}/{}/{}".format(month,date,year))
df['New_DOB']=df["DOB"].map(convert_date)
```

```python
def revised_ssn(ssn_string):
    return "xxxx-xx-" + ssn_string[-4:]
    
```

```python
df['New_SSN'] = df['SSN'].map(revised_ssn)
```
